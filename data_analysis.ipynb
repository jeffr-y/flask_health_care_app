```python
# Load data from MongoDB and write CSV
# This script loops through all stored users, converts them into objects,and them to a CSV file

import pandas as pd
from pymongo import MongoClient
import csv
from models.user_class import User 
import os

print("Current folder:", os.getcwd())

# ---- Connect to MongoDB ------
client = MongoClient("mongodb://localhost:27017/")
db = client["flask_health_care_app"]
users_collection = db["users"]

# -----  Check if there is data -----
if users_collection.count_documents({}) == 0:
    print("No document found in MongoDB")
else:
    # Fetch all documents
    records = users_collection.find()

    # Convert to User objects
    users = []
    for r in records:
        user = User(
            age=r["age"],
            gender=r["gender"],
            total_income=r["total_income"],
            expenses=r["expenses"]
        )
        users.append(user)

    # ---- Write to CSV ------
    csv_path = r"C:\Users\jeffr\flask_health_care_app\users_data.csv"
    with open(csv_path, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=users[0].to_dict().keys())
        writer.writeheader()
        for u in users:
            writer.writerow(u.to_dict())
    
    print(f"CSV saved at: {csv_path}")
    
    # -- Load CSV into pandas for verification
    df = pd.read_csv(csv_path)
    print(df.head())
    
```
