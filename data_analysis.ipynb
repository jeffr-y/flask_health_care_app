```python
# Load data from MongoDB and write CSV
# This script loops through all stored users, converts them into objects,and them to a CSV file

import os
import pandas as pd
from pymongo import MongoClient
from dotenv import load_dotenv
from models.user_class import User  # matches folder struture


print("Current folder:", os.getcwd())

# ---- Connect to MongoDB ------
load_dotenv()
client = MongoClient(os.getenv("MONGO_URI"))
db = client["flask_health_care_app"]
users_collection = db["users"]

# -----  Fetch all documents-----

records = list(users_collection.find({}, {"_id": 0}))

if not records:
    print("No document found in MongoDB")
else:
    # Convert MongoDB document - User objects
    
    users = []
    for r in records:
        user = User(
            age=r.get("age"),
            gender=r.get("gender"),
            total_income=r.get("total_income"),
            expenses=r.get("expenses")
        )
        users.append(user)
    #convert User objects - list of dictionaries
    user_dicts = [u.to_dict() for u in users]

    # Write CSV using Pandas

    df = pd.DataFrame(user_dicts)
    df.to_csv("data/user_data.csv", index=False)

    print("CSV written to data/user_data.csv")
    print(df.head())
    
   
    
```
